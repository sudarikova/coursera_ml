```python
import numpy as np
```

```python
#генерация случайной матрицы
```

```python
x = np.random.normal(loc=1, scale=10, size = (1000,50))
```

```python
print (x)
```

```python
#нормировка: вычитание из каждого столбца его ср.значения, деление на его станд.отклонение
```

```python
m = np.mean(x, axis=0)
std = np.std(x, axis=0)
x_norm = ((x-m)/std)
```

```python
print (x_norm)
```

```python
#номера строк матрицы, сумма эл-тов которых больше 10
```

```python
l = np.sum(x, axis=1)
print(np.nonzero(l>10))
```

```python
z = np.array([[4,5,0],
            [1,9,3],
            [5,1,1],
            [3,3,3],
            [9,9,9],
            [4,7,1]])
```

```python
r = np.sum(z, axis=1)
print(np.nonzero(r>10))
```
